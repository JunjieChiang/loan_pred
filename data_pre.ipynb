```python
import re
import os

import pandas as pd
import numpy as np
```

```python
root = os.getcwd()
print(root)

accept_data_path = os.path.join(root, "loan.csv")
accept_data = pd.read_csv(accept_data_path)
accept_data.info()
accept_data.head()
```

```python
'''只考虑loan_status的类别为Fully Paid和Default的数据'''
filtered_data = accept_data[accept_data['loan_status'].isin(["Fully Paid", "Default"])]
```

```python
filtered_data.info()
```

```python
'''清除desc中的无用描述'''
def clean_desc(desc):
    if isinstance(desc, str):
        return re.sub(r'Borrower added on \d{2}/\d{2}/\d{2} > ', '', desc)
    return desc

def clean_desc_2(desc):
    if isinstance(desc, str):
        return re.sub(r'<br>', '', desc)
    return desc

filtered_data['desc'] = filtered_data['desc'].apply(clean_desc)
filtered_data['desc'] = filtered_data['desc'].apply(clean_desc_2)
```

```python
filtered_data_path = os.path.join(root, "filtered_accepted_loans(FINAL).csv")
filtered_data.to_csv(filtered_data_path, index=False)
```
